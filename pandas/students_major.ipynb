```python
import pandas as pd

df = pd.read_csv("students_with_major_30.csv")

df.head()
```

```python
df.groupby("major")["score"].mean().sort_values(ascending=False)
```

```python
df["major"].value_counts()
```

```python
avg = df.groupby("major")["score"].mean()

# Ambil jurusan dengan rata-rata tertinggi
max_major = avg.idxmax()
max_value = avg.max()

print("Jurusan dengan nilai rata-rata tertinggi:", max_major)
print("Nilai rata-ratanya:", max_value)
```

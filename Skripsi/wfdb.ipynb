```python
import wfdb
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
record = wfdb.rdrecord("bidmc01")
```

```python
ppg  = record.p_signal[:, 0]  # kolom pertama
resp = record.p_signal[:, 1]  # kolom kedua

fs = record.fs  # 125 Hz
print("Sampling rate:", fs)
```

```python
plt.figure(figsize=(12,3))
plt.plot(ppg[:2000])
plt.title("PPG (window pertama)")
plt.grid(True)
plt.show()

plt.figure(figsize=(12,3))
plt.plot(resp[:2000])
plt.title("Respiration belt")
plt.grid(True)
plt.show()
```

```python
df = pd.DataFrame({
    "PPG": ppg,
    "RESP": resp
})
df.to_csv("bidmc01.csv", index=False)
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def __str__(self):
        return f"My name is {self.name} and my age is {self.age} years old"
    #this str function inside a class is used so that we can print the class only without 
    #referencing any properties. it needs a return value
person = Person("John", 36)
print(person)
```

```python
class Student(Person): #this is inheritance. the student class has all the properties of Person
    def __init__(self, name, age, year): # this makes the child class have its own constructor wihtout overriding the parent
        super().__init__(name, age)
        self.graduation_year = year # you can add more properties. dont forget to add it to the init parameters
    def __str__(self):
        return f"I am {self.name} and I am {self.age} years old. I go to university and will be graduating in {self.graduation_year}"

student = Student("Gerni", 20, 2026)
print(student)
```

```python
class MyNumber:
    def __iter__(self):
        self.a = 1
        return self
    def __next__(self):
        if self.a <= 20:
            x = self.a
            self.a += 1
            return x
        else:
            raise StopIteration

num = MyNumber()
newIter = iter(num)

for x in newIter:
    print(x)
```

```python
class Car:
    def __init__(self, brand, year):
        self.brand = brand
        self.year = year
    def move(self):
        print('Drive')

class Boat:
    def __init__(self, brand, year):
        self.brand = brand
        self.year = year
    def move(self):
        print('Sail')

class Plane:
    def __init__(self, brand, year):
        self.brand = brand
        self.year = year
    def move(self):
        print('Fly')
        
car = Car('Avansa', 2020)
boat = Boat('Yamaha', 2011)
plane = Plane('Garuda', 2010)

for x in (car, boat, plane):
    x.move() # polymorphism: the ability to access a method in several classes with the same name
    #with a for loop
```

```python
import matplotlib.pyplot as plt
import numpy as np

# there will be three points
x = np.array([2, 10, 1]) # this will show the x coordinates
y = np.array([3, 9, 7]) # these are the match y values

plt.plot(x, y, 'o') # without line (dont use '.' if you want a straight line)
plt.show()
```

```python
plt.plot(x, y, marker='o') # makes the line while also showing the bigger dots
plt.show()
```

```python
plt.plot(x,y, 'o--r') # marker|line|color. in this case o|dashed|red
plt.show()
```

```python
plt.plot(x,y, marker='o', ms=20, mec='r', mfc='g', ls='dashed') 
# ms is for the marker size
# mec is the outline of the marker
# mfc is the inline of the marker
# ls is the linestyle of the line

font1 = {
    'family': 'serif',
    'color': 'green',
    'size': 20
}
# font dictionary ^ v
font2 = {
    'family': 'serif',
    'color': 'red',
    'size': 30
}

plt.xlabel('X axis', fontdict=font1)
plt.ylabel('Y axis', fontdict=font1) # you know what this does. you just forget a lot
plt.title('ambasing', fontdict=font2, loc='right') # loc to set the position of the title

plt.grid(axis='x') # you also know what this does
plt.show()
```

# Pipelines

pipelines are used to combine preprocessing dan model training agar semuanya menjadi rapi, otomatis, dan aman dari data leakage.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

pipe = make_pipeline(
    StandardScaler(),
    LogisticRegression()
)
X, y = load_iris(return_X_y=True)
X_train, X_test, Y_train, Y_test = train_test_split(X, y, random_state=0)

pipe.fit(X_train, Y_train)
```

```python
accuracy_score(pipe.predict(X_test), Y_test)
```

# steps for using pipeline
- import the libraries like the estimator, StandardScaler, make_pipeline, accuracy_score, train_test_split
- create the pipeline
- split training with testing
- standardize it
- train
- test
- count accuracy

note: random_state=0 means to have the data randomized in one way only. Thus making the result the same everytime it runs. Without it, it will normally have different results all the time.

```python
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_validate

X, y = make_regression(n_samples=1000, random_state=0)
lr = LinearRegression()

result = cross_validate(lr, X, y)
result['test_score']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np
X, y = make_regression(n_samples=1000, n_features=1, random_state=0)

reg = LinearRegression()
reg.fit(X, y)

plt.scatter(X, y, color='blue', label='Data asli')  # titik data
plt.plot(X, reg.predict(X), color='red', label='Garis regresi')  # garis lurus
plt.xlabel("Ukuran Rumah (dalam satuan)")
plt.ylabel("Harga Rumah")
plt.title("Regresi Linear Sederhana")
plt.legend()
plt.grid(True)
plt.show()
```

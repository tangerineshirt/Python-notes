```python
inputs = [1,2,3,2.5]
weights = [[0.2, 0.8, -0.5, 1.0],
            [0.5, -0.91, 0.26, -0.5],
            [-0.26, -0.27, 0.17, 0.87]]

bias = [2, 3, 0.5]

output = [0,0,0]

for i in range(len(weights)):
    for j in range(len(inputs)):
        output[i] += inputs[j]*weights[i][j]
    output[i] += bias[i]

print(output)
```

```python
inputs = [1,2,3,2.5]
weights = [[0.2, 0.8, -0.5, 1.0],
            [0.5, -0.91, 0.26, -0.5],
            [-0.26, -0.27, 0.17, 0.87]]

bias = [2,3,0.5]
output = []
j=0
for neuron_weights in weights:
    temp = 0
    for i, w in zip(neuron_weights, inputs):
        temp += i*w
    temp += bias[j]
    j += 1
    output.append(temp)

print(output)
```

```python
import numpy as np
inputs = np.array([1,2,3,2.5])
weights = np.array([
    [0.2,0.8,-0.5,1.0],
    [0.5,-0.91,0.26,-0.5],
    [-0.26,-0.27,0.17,0.87]
])

biases = np.array([2,3,0.5])

output = np.array([])

for neuron_weights, bias in zip(weights, biases):
    temp = np.dot(inputs, neuron_weights)
    temp += bias
    temp_arr = np.append(output, temp)
    output = temp_arr

print(output)
```

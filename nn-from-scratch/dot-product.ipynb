```python
inputs = [1,2,3,2.5]
weights = [[0.2, 0.8, -0.5, 1.0],
            [0.5, -0.91, 0.26, -0.5],
            [-0.26, -0.27, 0.17, 0.87]]

bias = [2,3,0.5]
output = []
j=0
for neuron_weights in weights:
    temp = 0
    for i, w in zip(neuron_weights, inputs):
        temp += i*w
    temp += bias[j]
    j += 1
    output.append(temp)

print(output)
```

```python
import numpy as np
inputs = np.array([1,2,3,2.5])
weights = np.array([
    [0.2,0.8,-0.5,1.0],
    [0.5,-0.91,0.26,-0.5],
    [-0.26,-0.27,0.17,0.87]
])

biases = np.array([2,3,0.5])

output = np.dot(weights, inputs) + biases

print(output)
print(output.shape)
```

```python
# PRACTICE FOR LIST COMPREHENSION
# square of numbers 1-5
array = np.array([i**2 for i in range(1,6)])
print(f"The square of the first five numbers are: {array}")

# even numbers until 20
even = np.array([x for x in range(21) if x%2==0])
print(f"All even numbers until 20: {even}")

# turn all strings into uppercase
strings = np.array(['hello', 'python', 'world'])
upper_strings = np.array([string.upper() for string in strings])
print(f"Uppercase of the strings array is {upper_strings}")

# combine two lists and then print them
names = ["Alice", "Bob", "Charlie"]
ages = [25, 30, 22]
combine = [f"{name}: {age}" for name, age in zip(names,ages)]
print(combine)

# nested list comprehension
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
nested = np.array([column**2 for row in matrix for column in row])
print(f"From 2D to 1D: {nested}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def function(x):
    return 2*x**2

x = np.arange(0,3, 0.1)
y = function(x)
plt.plot(x, y, color="red")

delta = 0.001
x1 = 2
x2 = x1+delta

y1 = function(x1)
y2 = function(x2)

approx_dx = (y2-y1) / (x2-x1)
b = y1 - approx_dx*x1

def tangent_line(x):
    return approx_dx*x + b

to_plot = np.array([x1-0.9, x1, x1+0.9])
graph = np.array([tangent_line(i) for i in to_plot])
plt.plot(to_plot, graph)

plt.grid(True)
plt.show()
```

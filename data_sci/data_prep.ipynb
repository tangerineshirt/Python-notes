```python
import pandas as pd
import numpy as np
```

```python
health_data = pd.read_csv("health_data.csv")
health_data.head()
```

```python
health_data = health_data.drop(columns='RiskLevel')
```

```python
health_data.dropna(axis=0, inplace=True)
```

```python
print(health_data.info())
```

```python
health_data['Age'] = health_data['Age'].astype('float64')
health_data['SystolicBP'] = health_data['SystolicBP'].astype('float64')
health_data['DiastolicBP'] = health_data['DiastolicBP'].astype('float64')
health_data['HeartRate'] = health_data['HeartRate'].astype('float64')
```

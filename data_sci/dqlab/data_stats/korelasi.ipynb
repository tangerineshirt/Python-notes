```python
import pandas as pd
import numpy as np
raw_data = pd.read_csv("https://storage.googleapis.com/dqlab-dataset/dataset_statistic.csv", sep=';')
```

```python
pendapatan = raw_data['Pendapatan']
harga = raw_data['Harga']
print(pendapatan)
```

```python
import matplotlib.pyplot as plt

plt.plot(pendapatan, harga)
plt.grid(True)
plt.title("Pendapatan x Harga")
plt.xlabel("Pendapatan")
plt.ylabel("Harga")
plt.show()
```

# korelasi pearson



```python
#manual way
n = len(pendapatan)
sum_xy = sum(pendapatan * harga)
sum_x = sum(pendapatan)
sum_y = sum(harga)
sum_x2 = sum(pendapatan**2)
sum_y2 = sum(harga**2)
pearson = (n*sum_xy - sum_x * sum_y) / (np.sqrt(n*sum_x2 - sum_x**2) * np.sqrt(n*sum_y2 - sum_y**2))
print(pearson)
```

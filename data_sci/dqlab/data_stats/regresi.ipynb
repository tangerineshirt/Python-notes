# contoh regresi sederhana
menggunakan statsmodel

```python
import pandas as pd
import statsmodels.api as sm

raw_data = pd.read_csv("https://storage.googleapis.com/dqlab-dataset/dataset_statistic.csv", sep=';')
print(raw_data)
```

```python
y = raw_data['Total']
x = sm.add_constant(raw_data[['Pendapatan']])
```

```python
model_regresi = sm.OLS(endog=y, exog=x).fit()
print(model_regresi.summary( ))
```

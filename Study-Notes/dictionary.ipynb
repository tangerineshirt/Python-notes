```python
diction = {
    "name": "Razzan Naufal Rianta",
    "age": 20,
    "matkuls": ["Embedded Systems", "Cloud", "Fault Tolerant"]
} # this is a dictionary

print(diction["name"]) # access the dictionary with square brackets

x = diction.get('matkuls') # also like this
print(x)
```

```python
print(diction.keys()) # gets all the keys of the dictionary
print(diction.values()) # gets all the values of the dictionary
```

```python
# updating the values
print(diction['name']) # before change
diction["name"] = "Test man"
print(diction['name']) # after change

# OR

diction.update({"age": 21})
print(diction["age"])

# hey neat trick below

put = "age"
diction.update({put: 30})
print(diction["age"])
```

```python
# adding new items
diction['newThing'] = "idk" # just like normal

print(diction['newThing'])
```

```python
# removing items
diction.pop("newThing") # removes a specific value
print(diction)
```

```python
# another way of removing items
diction.popitem() # removes the last item in the dictionary
print(diction)
```

```python
# delete a dictionary completely
del diction
```

```python
# clearing a dictionary

thisdict = {
    "thing": "Hello world",
    "otherThing": 70
}

print(thisdict)
```

```python
# looping in a dictionary
print("Keys in a dictionary: ")
for x in thisdict:
    print(x)

print("\nValues in a dictionary: ")
for x in thisdict:
    print(thisdict[x])

print("\nValues in a dictionary (other way): ")
for x in thisdict.values():
    print(x)

print("\nLoop through the keys and values: ")
for x, y in thisdict.items():
    print(x, y)
```

```python
# copy
newdict = thisdict.copy()
print(newdict)
```

```python
child1 = {
    "name": "Siti",
    "age": 10
}

child2 = {
    "name": "Budi",
    "age": 20
}

child3 = {
    "name": "Udin",
    "age": 15
}

parent = {
    "child1": child1,
    "child2": child2,
    "child3": child3
}

for key, val in parent.items():
    print(f"\nThis is {key}")
    for x in val:
        print(f"{x}: {val[x]}")
```

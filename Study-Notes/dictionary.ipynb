```python
# clearing a dictionary

thisdict = {
    "thing": "Hello world",
    "otherThing": 70
}

print(thisdict)
```

```python
# looping in a dictionary
print("Keys in a dictionary: ")
for x in thisdict:
    print(x)

print("\nValues in a dictionary: ")
for x in thisdict:
    print(thisdict[x])

print("\nValues in a dictionary (other way): ")
for x in thisdict.values():
    print(x)

print("\nLoop through the keys and values: ")
for x, y in thisdict.items():
    print(x, y)
```

# references

in python, instead of pointers there are references. it is just using an equals to sign. that's it. BUT it only works for mutable objects like dictionaries, list, tuples, object instances of classes, etc. immutable objects like int, float, bool, string, etc, cannot be referenced. lets try it shall we

```python
class Node:
    def __init__(self, data, right=None, left=None):
        self.data = data
        self.right = right
        self.left = left

class BinaryTree:
    def __init__(self):
        self.root = None
        self.size = 0
    
    def __sisipDt(self, input, node):
        if self.root == None:
            self.root = Node(input)
            self.size += 1
        else:
            if input >= node.data:
                if node.right == None:
                    node.right = Node(input)
                    self.size += 1
                else:
                    self.__sisipDt(input, node.right)
            elif input < node.data:
                if node.left == None:
                    node.left = Node(input)
                    self.size += 1
                else:
                    self.__sisipDt(input, node.left)
    
    def __preorderDt(self, node):
        if node is None:
            return 
        print(node.data)
        self.__preorderDt(node.left)
        self.__preorderDt(node.right)
    
    def __inorderDt(self, node):
        if node is None:
            return 
        self.__inorderDt(node.left)
        print(node.data)
        self.__inorderDt(node.right)
    
    def __postorderDt(self, node):
        if node is None:
            return 
        self.__postorderDt(node.left)
        self.__postorderDt(node.right)
        print(node.data)
    
    def sisipData(self, input):
        self.__sisipDt(input, self.root)
    
    def preorder(self):
        self.__preorderDt(self.root)
    
    def inorder(self):
        self.__inorderDt(self.root)
    
    def postorder(self):
        self.__postorderDt(self.root)

tree = BinaryTree()

tree.sisipData(30)
tree.sisipData(40)
tree.sisipData(5)
tree.sisipData(10)
tree.sisipData(15)
tree.inorder()
```

```python
import numpy as np

np.random.seed(0)

X = np.array([
    [1,2.5,-3,1],
    [4,-9,-1,6],
    [6,-5,-2,-7]
])

class Layer_Dense:
    def __init__(self, n_input, n_neuron):
        self.weights = 0.2 * np.random.randn(n_input, n_neuron)
        self.biases = np.zeros((1, n_neuron))
    def forward(self, input):
        self.output = np.dot(input, self.weights) + self.biases

layer1 = Layer_Dense(4, 6)
layer1.forward(X)
print(layer1.output)
```
